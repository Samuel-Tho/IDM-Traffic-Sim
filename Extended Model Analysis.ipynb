```python
aa = []
bb = []
cc = []
x = np.linspace(0,0.05,100)
for i in range(100):
    aa.append(stat.mean(unhap_carson_j[0,i]))
    bb.append(stat.mean(unhap_carson_jtl[0,i]))
    cc.append(stat.mean(unhap_carson_jtlg[0,i]))
plt.plot(x,aa,label = 'Junction Only')
plt.plot(x,bb, label = 'Junction With Traffic Light')
plt.plot(x,cc, label = 'Junction With Give Way And Traffic Light')
plt.ylabel('Average length of que')
plt.xlabel('Probability that a car arrives at junction every 0.5 seconds')
plt.legend()
```

```python
base_blue = (0, 0, 1)  # RGB values for blue (0, 0, 1)
num_shades = 3
blue_array = [base_blue]
for i in range(1, num_shades):
    darker_blue = tuple(max(0, c - 0.35 * i) for c in base_blue)  # Adjust darkness
    blue_array.append(darker_blue)
    
base_red = (1, 0, 0)  # RGB values for red (1, 0, 0)
num_shades = 3
red_array = [base_red]
for i in range(1, num_shades):
    darker_red = tuple(max(0, c - 0.35 * i) for c in base_red)  # Adjust darkness
    red_array.append(darker_red)

plt.plot(x,b_fdp_j[0,:],label = 'Junction Only, 10 Metres Before',color = blue_array[0])
plt.plot(x,b_fdp_jtl[0,:],label = 'Junction With Traffic Light, 10 Metres Before',color = red_array[0])
plt.plot(x,a_fdp_j[0,:],label = 'Junction Only, 10 Metres After',color = blue_array[1])
plt.plot(x,a_fdp_jtl[0,:],label = 'Junction With Traffic Light, 10 Metres After',color = red_array[1])
plt.plot(x,g_fdp_j[0,:],label = 'Junction Only, 1000 Metres After',color = blue_array[2])
plt.plot(x,g_fdp_jtl[0,:],label = 'Junction With Traffic Light, 1000 Metres After',color = red_array[2])
#plt.plot(x,b_fdp_jtlg[0,:],label = 'Junction With Give Way And Traffic Light')
plt.xlabel('Probability that a car arrives at junction every 0.5 seconds')
plt.ylabel('Average local flow')
plt.title('Local Flow At Detection Points')
plt.legend()
plt.savefig('Local Flow Overall',dpi = 500)
plt.show()
```
